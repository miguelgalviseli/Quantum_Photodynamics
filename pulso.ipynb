```python
import numpy as np
import matplotlib.pyplot as plt
E0 = 1
omega_l = 0.05
tf = 2000

def pulso(t,):
    return E0*(np.sin(np.pi*t/tf)**2) *np.cos(omega_l*(t-tf/2))

def envolvente(t):
    return E0*(np.sin(np.pi*t/tf)**2)

plt.figure(figsize=(10,5))
t = np.linspace(0,tf,1000)
plt.plot(t,pulso(t), color='yellowgreen', linewidth=2.5, label='Pulso')
plt.plot(t,envolvente(t), color='black', linewidth=3.5, label='Envolvente')
plt.xlabel('t',fontsize=20)
plt.tick_params(labelsize=20)
plt.ylabel('E(t)',fontsize=25)  
plt.grid()
plt.legend(fontsize=20)
plt.show()

```

```python
#Queremos ver un area sombreada hasta la mitad del pulso
# Para eso vamos a usar la funcion fill_between

plt.figure(figsize=(10,5))
#plt.plot(t,pulso(t))
plt.plot(t,envolvente(t), color='black', linewidth=3.5)
plt.xlabel('t',fontsize=20)
plt.ylabel(r'$E(t)$',fontsize=25)
plt.fill_between(t,envolvente(t),where=t<3*tf/4,color='yellowgreen',alpha=0.5)
#Font de los sticks
plt.tick_params(labelsize=20)

#plt.grid()
plt.show()
```
